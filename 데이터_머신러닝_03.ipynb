```python
import statsmodels.formula.api as smf

# 데이터프레임(df)에 'price', 'mileage', 'year' 등의 변수가 있어야 합니다.

# 회귀 모델을 정의합니다.
model = smf.ols('price ~ scale(mileage) + scale(year)', data=df)

# 모델을 적합합니다.
results = model.fit()

# 모델 결과 요약을 출력합니다.
print(results.summary())
```

```python
m = smf.ols('price ~ scale(mileage) + scale(year)', data=df).fit()
m.summary()
```

price = 853.6606 -100 scale(mileage) + 218 scale(year)
* 표준화했을 때 절편이 853만원이라는 뜻:
평균적인 mileage와 평균적인 year의 차가
853만원 정도 한다
* mileage가 1표준편차(43015Km) 증가하면
가격은 100만원 떨어진다
* year가 1표준편차(2.6년) 증가하면 가격은 218만원 상승
* year가 가격에는 영향이 더 크다고 할 수 있음

```python
m = smf.ols('price ~ model', data=df).fit()
m.summary()
```

- Intercept	833.4146 (Avante 예상가격)
- model[T.K3]	80.3970 (K3와 Avante의 예상가격 차이)

`price = 833 + 80 * model[T.K3]`

- Avante: `model[T.K3] = 0` 833만원으로 예상
- K3: `model[T.K3] = 1` 913만원으로 예상

- p-value > 0.05이므로 귀무가설: "둘의 차이 = 0"을 기각할 수 없음
- 또는 신뢰구간 -10~+171(0도 포함됨)
- 따라서 모집단에서도 K3가 Avante보다 비싸다고 하기에는 증거가 부족하다






```python
avante = df.price[df.model == 'Avante']
k3 = df.price[df.model == 'K3']
pg.ttest(k3, avante, correction=False)
#독립표본 t검정에서 보정을 안하면 회귀분석과 결과 같음
#무엇을 보정? 이분산성 : 가격대가 퍼져있는 정도가 다르다
```
